```python
import pandas as pd
import psycopg2
```

```python
# Connect to PostgreSQL using NEON_DATABASE_URL (do NOT hardcode credentials)
try:
    from dotenv import load_dotenv
    load_dotenv()
except Exception:
    pass

import os
import psycopg2

connection_url = os.environ.get('NEON_DATABASE_URL')
if not connection_url:
    raise RuntimeError("NEON_DATABASE_URL is not set. Add it to your environment or .env and re-run this cell.")

# psycopg2 accepts a connection string (DSN)
conn = psycopg2.connect(connection_url)
```

```python
# Define a Helper Function to Run SQL and Return DataFrame

def run_query(sql):
    return pd.read_sql(sql, conn)
```

```python
# Show all reviews
sql = "SELECT * FROM reviews;"

run_query(sql)
```

```python
# Show all banks
sql = "SELECT * FROM banks;"

run_query(sql)
```

```python
# Count reviews per bank
sql = """
SELECT b.bank_name, COUNT(*) AS total_reviews
FROM reviews r
JOIN banks b ON b.bank_id = r.bank_id
GROUP BY b.bank_name
ORDER BY total_reviews DESC;
"""

run_query(sql)
```

```python
# Average rating per bank
sql = """
SELECT b.bank_name, ROUND(AVG(r.rating), 2) AS avg_rating
FROM reviews r
JOIN banks b ON b.bank_id = r.bank_id
GROUP BY b.bank_name
ORDER BY avg_rating DESC;
"""

run_query(sql)
```
